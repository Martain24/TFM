```python
df_train = pd.read_csv("Train.csv").drop("ID", axis="columns")
df_train
```

```python
df_train.isna().sum()
```

```python
df_train = df_train.dropna().reset_index(drop=True)
```

```python
X_variables = ["Gender", "Ever_Married", "Age", "Graduated", "Profession", "Work_Experience"]
y_variable = "Spending_Score"
```

```python
from sklearn.linear_model import LogisticRegression
X = pd.get_dummies(df_train[X_variables])
y = df_train[y_variable]

model = LogisticRegression(max_iter=5000).fit(X, y)
```

```python
from sklearn.metrics import balanced_accuracy_score
y_train_pred = model.predict(X)
balanced_accuracy_score(y_pred=y_train_pred, y_true=y)
```
